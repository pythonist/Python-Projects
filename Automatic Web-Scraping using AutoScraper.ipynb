```python
from autoscraper import AutoScraper
```

```python
#initializing the url we want to scrap and the list of items
url = "https://www.flipkart.com/search?q=graphics+card"
w_list = ["₹18,999","ASUS NVIDIA Cerberus GeForce GTX 1050 Ti 4GB OC Edition..."]
```

```python
#creating scraper object
scraper = AutoScraper()
result = scraper.build(url,w_list)
print(result)
```

```python
#grouping the results
scraper.get_result_similar(url,grouped = True)
```

```python
#saving the rules as title and price
scraper.set_rule_aliases({'rule_da4u':'Title','rule_5ctw':'Price'})
scraper.keep_rules(['rule_da4u','rule_5ctw'])
scraper.save('GPU Search')
```

```python
#using similar rules to scrap a new page url
result = scraper.get_result_similar('https://www.flipkart.com/search?q=iphones',group_by_alias = True)
```
